```python
# import dependencies
import requests as req
import pandas as pd
```

```python
# NYT API query parameters
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json"
api_key = "d673bd0cbe3f44d18b957ffaf3694170"
q = 'central african republic',
begin_date = "20150101",
end_date = "20171231",
page = 0
```

```python
# set param_dict
param_dict = {
    'api-key': api_key,
    'q': q,
    'begin_date': begin_date,
    'end_date': end_date,
    'page':page
}
```

```python
# retrieve API results
api_body = req.get(url, params=param_dict).json()
```

```python
# print meta of api_body response to get hits (number of articles queryed) Norway: 150, Denmark: 150, Burundi: 35, CAR:44 
param_dict['q'] = 'central african republic'
api_body = req.get(url, params=param_dict).json()
pages = int(api_body['response']['meta']['hits']/10)
print(pages)
```

```python
# print headline and snippet from api response
for article in api_body['response']['docs']:
    print("Headline: '" + article["headline"]["main"] + "'.\n"\
          "Snippet: '" + article["snippet"] + "'.\n"\
          "Date: '" + article["pub_date"])
```

```python
# set variables to empty list
headlines = []
snippets = []
dates = []
# loop through pages and append variables to lists
for api_page in range(0,pages):
    param_dict['page'] = api_page
    for article in api_body['response']['docs']:
        headlines.append(article["headline"]["main"])
        snippets.append(article["snippet"])
        dates.append(article["pub_date"])
        #articles.append("Headline: " + article["headline"]["main"])
        #articles.append("Snippet: " + article["snippet"])
headlines
```

```python
# put lists into df
CAR_NYT_Data = pd.DataFrame(
    {'Headlines': headlines,
     'Snippets': snippets,
     'Year': dates
    })

CAR_NYT_Data.dtypes
```

```python
# get year from date
CAR_NYT_Data['Year'] = CAR_NYT_Data.Year.str[:4]
CAR_NYT_Data['Country'] = 'CAR'
CAR_NYT_Data.head()
```

```python
CAR_NYT_Data.to_csv("CAR_NYT_Data.csv")
```
