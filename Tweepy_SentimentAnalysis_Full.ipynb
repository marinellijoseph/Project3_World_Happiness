## Denmark

```python

# Lists to hold sentiments
compound_list = []
positive_list = []
negative_list = []
neutral_list = []

# Grab 1000 tweets
public_tweets = api.search(count=1000, geocode="55.68028,12.59005,15000km", result_type="recent")

# Loop through all tweets
for tweet in public_tweets["statuses"]:

    # Run Vader Analysis on each tweet
    compound = analyzer.polarity_scores(tweet["text"])["compound"]
    pos = analyzer.polarity_scores(tweet["text"])["pos"]
    neu = analyzer.polarity_scores(tweet["text"])["neu"]
    neg = analyzer.polarity_scores(tweet["text"])["neg"]

    # Add each value to the appropriate array
    compound_list.append(compound)
    positive_list.append(pos)
    negative_list.append(neg)
    neutral_list.append(neu)
    
sentiment_Denmark = {"Compound": np.mean(compound_list),
             "Positive": np.mean(positive_list),
             "Neutral": np.mean(negative_list),
             "Negative": np.mean(neutral_list)}

# Print the Sentiments

#print(tweet["created_at"],tweet["text"], tweet["lang"])
print(sentiment_Denmark)
```

```python
#Print Data Frame
Denmark_pd = pd.DataFrame(sentiment_Denmark, index = [0])
Denmark_pd["Country"] = ["Denmark"]
Denmark_pd
```
