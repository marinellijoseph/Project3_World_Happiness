```python
import pandas as pd
```

```python
file_one = "Data/Happiness/2015_Happy.csv"
file_two = "Data/Happiness/2016_Happy.csv"
file_three = "Data/Happiness/2017_Happy.csv"
```

```python
file_one_df = pd.read_csv(file_one)
#file_one_df.head()
```

```python
year15 = [2015]*158
file_one_df['year'] = year15
#list(file_one_df.columns)
```

```python
file_two_df = pd.read_csv(file_two)
#list(file_two_df.columns)
```

```python
year16 = [2016]*157
file_two_df['year'] = year16
#file_two_df.head()
```

```python
file_three_df = pd.read_csv(file_three)
#list(file_three_df.columns)
```

```python
year17 = [2017]*155
file_three_df['year'] = year17
file_three_df = file_three_df.rename(columns={'Dystopia.Residual':'Dystopia Residual','Economy..GDP.per.Capita.':'Economy (GDP per Capita)', 
                                                  'Happiness.Rank':'Happiness Rank',
                                                  'Happiness.Score': 'Happiness Score', 
                                                  'Health..Life.Expectancy.':'Health (Life Expectancy)', 
                                                  'Economy..GDP.per.Capita.':'Economy (GDP per Capita)', 
                                                  'Trust..Government.Corruption.':'Trust (Government Corruption)'
                                                 })
#file_three_df.head()
```

```python
frames = [file_one_df, file_two_df, file_three_df]

Happiness_df = pd.concat(frames, join = 'inner')
#list(Happiness_df.columns)
Happiness_df.head()
```

```python
file_corruptionone = "Data/Corruption/CPI_2015_FullDataSet.csv"
file_corruptiontwo = "Data/Corruption/CPI_2016_FullDataSet.csv"
file_corruptionthree = "Data/Corruption/CPI_2017_FullDataSet.csv"
```

```python
file_corruptionone_df = pd.read_csv(file_corruptionone, encoding = "ISO-8859-1")
file_corruptiontwo_df = pd.read_csv(file_corruptiontwo, encoding = "ISO-8859-1")
file_corruptionthree_df = pd.read_csv(file_corruptionthree, encoding = "ISO-8859-1", skiprows=2)
```

```python
year_15 = [2015]*168
year_16 = [2016]*176
year_17 = [2017]*182
file_corruptionone_df['year'] = year_15
file_corruptiontwo_df['year'] = year_16
file_corruptionthree_df['year'] = year_17

file_corruptionone_df.columns
file_corruptiontwo_df.columns
file_corruptionthree_df.columns
```

```python
file_corruptionone_df = file_corruptionone_df.rename(columns={'Country/Territory':'Country',
                                                                  'CPI 2015 Score':'CPI'
                                                                 })
file_corruptiontwo_df = file_corruptiontwo_df.rename(columns={'CPI2016':'CPI'
                                                                 })
file_corruptionthree_df = file_corruptionthree_df.rename(columns={'CPI Score 2017':'CPI',
                                                                  'Region':'reg'
                                                                 })
file_corruptionone_df.head()
file_corruptiontwo_df.head()
file_corruptionthree_df.head()
```

```python
frames1 = [file_corruptionone_df, file_corruptiontwo_df, file_corruptionthree_df]

Corruption_df = pd.concat(frames1, join = 'inner')
Corruption_df.head()
```

```python
#frames2 = [Corruption_df, Happiness_df]
Happiness_corruption_df = pd.merge(Happiness_df, Corruption_df, how = 'inner', on = ['Country', 'year'])
Happiness_corruption_df
```
